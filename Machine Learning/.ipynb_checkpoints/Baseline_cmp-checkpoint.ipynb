```python
from sklearn.linear_model import LogisticRegression, LogisticRegressionCV
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, RobustScaler, LabelEncoder, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.metrics import roc_auc_score, roc_curve, make_scorer, plot_roc_curve
```

### Делим датафрейм на обучающую и тестувую части

```python
data_st = data.copy() # Скопируем для эксперементов
```

Ранее отмечалось, что наблюдались "выбросы" на боксплотах для Churn = 1.
Попробуем их исключить из датафрйма.
<br>P.s. Это действиетльно повысило точность оценки, поэтому оставляем

```python
data_st = data_st.drop(np.where((data_st['ClientPeriod'] > 69.5) & (data_st['Churn'] == 1))[0])
```

```python
num_cols_for_tr = num_cols.copy() # Числовые признаки, по которм будет учиться и работать модель
# num_cols_for_tr.remove('ClientPeriod') # Удаление одного из коррелирующих признаков дало худший результат
cat_cols_for_tr = cat_cols.copy() # Кат. признаки, по которм будет учиться и работать модель
```

```python
num_cols_for_tr
```

```python
x_data = data_st[num_cols_for_tr + cat_cols_for_tr]
y_data = data_st[target_col]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data.values, train_size=0.8, random_state=83)
```

### LogisticRegressionCV

#### OneHot-кодируем категоральные признаки и нормируем

```python
# OneHot-кодируем
x_train_lrcv = pd.concat([x_train[num_cols_for_tr], 
                    pd.get_dummies(x_train[cat_cols_for_tr], drop_first=True)], 
                   axis=1)
x_test_lrcv = pd.concat([x_test[num_cols_for_tr], 
                    pd.get_dummies(x_test[cat_cols_for_tr], drop_first=True)], 
                   axis=1)
# Нормируем
scaler = StandardScaler()
x_train_lrcv_sc = scaler.fit_transform(x_train_lrcv)
x_test_lrcv_sc = scaler.transform(x_test_lrcv)
```

#### Обучаем LogisticRegressionCV

```python
# Обучаем
lr_cv = LogisticRegressionCV(Cs=[100, 10, 1, 0.1, 0.01, 0.001], refit=True, 
                             scoring='roc_auc', cv=5, 
                             solver='lbfgs', max_iter=500)
res = lr_cv.fit(x_train_lrcv_sc, y_train)
```

```python
# Определяем вероятностный результат работы модели
y_lrcv_pred_trian = lr_cv.predict_proba(x_train_lrcv_sc)[:, 1]
y_lrcv_pred_test = lr_cv.predict_proba(x_test_lrcv_sc)[:, 1]
```

```python
# Строим ROC-AUC
train_auc = roc_auc_score(y_train, y_lrcv_pred_trian)
test_auc = roc_auc_score(y_test, y_lrcv_pred_test)
plt.figure(figsize=(20,10))
plt.plot(*roc_curve(y_train, y_lrcv_pred_trian)[:2], label='train AUC={:.4f}'.format(train_auc))
plt.plot(*roc_curve(y_test, y_lrcv_pred_test)[:2], label='test AUC={:.4f}'.format(test_auc))
legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
legend_box.set_facecolor("white")
legend_box.set_edgecolor("black")
plt.plot(np.linspace(0,1,100), np.linspace(0,1,100))
plt.show()
```

```python
# Метрика 'roc-auc' для разных 'C' на каждом шаге cv
lr_cv.scores_
```

```python
# Наилучшее значение С
lr_cv.C_
```

### LogisticRegression и GridSearchCV

```python
from sklearn.compose import ColumnTransformer, make_column_transformer
```

```python
tun_param_lr = {'logisticregression__C': [100, 10, 1, 0.1], 
            'logisticregression__class_weight': [None, 'balanced'], 
            'logisticregression__solver': ['newton-cg', 'lbfgs', 'liblinear']}

cat_transform = make_column_transformer((OneHotEncoder(drop='first'), cat_cols), 
                                       remainder='passthrough')

lr_cv_custom = make_pipeline(cat_transform, StandardScaler(), LogisticRegression(max_iter=500))
lr_cv_grid = GridSearchCV(lr_cv_custom, param_grid=tun_param_lr, scoring='roc_auc', cv=5, verbose=1)
lr_cv_grid.fit(x_train, y_train)
```

```python
# Определяем вероятностный результат работы модели
y_lrcv_cust_pred_trian = lr_cv_grid.predict_proba(x_train)[:, 1]
y_lrcv_cust_pred_test = lr_cv_grid.predict_proba(x_test)[:, 1]
```

```python
# Строим ROC-AUC
train_auc = roc_auc_score(y_train, y_lrcv_cust_pred_trian)
test_auc = roc_auc_score(y_test, y_lrcv_cust_pred_test)
plt.figure(figsize=(20,10))
plt.plot(*roc_curve(y_train, y_lrcv_cust_pred_trian)[:2], label='train AUC={:.4f}'.format(train_auc))
plt.plot(*roc_curve(y_test, y_lrcv_cust_pred_test)[:2], label='test AUC={:.4f}'.format(test_auc))
legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
legend_box.set_facecolor("white")
legend_box.set_edgecolor("black")
plt.plot(np.linspace(0,1,100), np.linspace(0,1,100))
plt.show()
```

##### Лучшее качество и параметры модели LogisticRegression, которые были получены:

```python
# Лучшая метрика на тестовых данных
roc_auc_score(y_test, y_lrcv_cust_pred_test)
```

```python
# Лучшие параметры, подобранные по сетке
lr_cv_grid.best_params_
```

```python
default_boosting_model = catboost.CatBoostClassifier(
                                             verbose=20, 
                                            cat_features=np.arange(3, x_train.shape[1]), custom_loss='AUC')
default_boosting_model.fit(x_train, y_train, eval_set=(x_test, y_test), plot=True)
```

### CatBoostClassifier с подобранными параметрами

```python
par = {
    'loss_function': 'Logloss',
    'iterations': 200,
    'custom_loss': 'AUC', 
    'eval_metric': 'AUC', 
    'early_stopping_rounds': 100,
    'learning_rate': 0.2,
    'use_best_model': True, 
    'bootstrap_type': 'Bayesian',
    'bagging_temperature': 1,
    'random_strength': 10,
    'thread_count': 4, 
    'depth': 6, 
    'l2_leaf_reg': 2
}

best_boosting_model = catboost.CatBoostClassifier(
                                             verbose=20, 
                                            cat_features=np.arange(3, x_train.shape[1]), 
                                            **par)
best_boosting_model.fit(x_train, y_train, eval_set=(x_test, y_test), plot=True)
```

```python
y_train_predicted = best_boosting_model.predict_proba(x_train)[:, 1]
y_test_predicted = best_boosting_model.predict_proba(x_test)[:, 1]
```

```python
train_auc = roc_auc_score(y_train, y_train_predicted)
test_auc = roc_auc_score(y_test, y_test_predicted)

plt.figure(figsize=(10,7))
plt.plot(*roc_curve(y_train, y_train_predicted)[:2], label='train AUC={:.4f}'.format(train_auc))
plt.plot(*roc_curve(y_test, y_test_predicted)[:2], label='test AUC={:.4f}'.format(test_auc))
legend_box = plt.legend(fontsize='large', framealpha=1).get_frame()
legend_box.set_facecolor("white")
legend_box.set_edgecolor("black")
plt.plot(np.linspace(0,1,100), np.linspace(0,1,100))
plt.show()
```
